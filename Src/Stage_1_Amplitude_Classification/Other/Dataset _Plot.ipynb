# CSV data plotting

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
%matplotlib inline

data = pd.read_csv('data_vis(6-directions).csv')

sns.FacetGrid(data,hue='DIRECTION',height=6)\
.map(plt.scatter,'RIGHT','LEFT')\
.add_legend()
```

```python
data = pd.read_csv('data_vis(4-directions).csv')

sns.FacetGrid(data,hue='DIRECTION',size=8)\
.map(plt.scatter,'RIGHT','LEFT')\
.add_legend()
```
